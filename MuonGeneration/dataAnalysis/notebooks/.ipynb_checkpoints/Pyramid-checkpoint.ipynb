```python
import ROOT as r
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors


fopen = r.TFile("../datos/output-Open.root")
fpyram = r.TFile("../datos/output-Pyramid.root")

topen = fopen.Get('events')
tpyram = fpyram.Get('events')

hopen = r.TH2F('hopen', '', 100, -4.0, 4.0, 100, 0, 4.0)
hpyram = r.TH2F('hpyram', '', 100, -4.0, 4.0, 100, 0, 4.0)

#for ev in topen:
#    r = np.sqrt(ev.vx**2+ev.vy**2+ev.vz**2)
#    phi = np.arctan2(ev.vy, ev.vx)
#    theta = np.arccos(ev.vz/r)
#    hopen.Fill(phi, theta)

phi = []
theta = []
for ev in tpyram:
    r = np.sqrt(ev.vx**2+ev.vy**2+ev.vz**2)
    phi.append(np.arctan2(ev.vy, ev.vx))
    theta.append(np.arccos(ev.vz/r))

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(10, 8) )

axs[0].set_title('Linear normalization')
axs[0].hist2d(phi, theta, bins=100, range=[[-4.0, 4.0], [0.0, 4.0]])

plt.show()
```
